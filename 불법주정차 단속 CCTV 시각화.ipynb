```python
layer = pdk.Layer(
    'PolygonLayer',
    data,
    get_polygon = 'coordinates',
    get_fill_color = '[0, 255*정규화인구, 0]',
    pickable=True,
    auto_highlight=True
)

center = [37.549593, 127.081780]
view_state =pdk.ViewState(
    longitude=center[0],
    latitude=center[1],
    zoom=12)
r=pdk.Deck(layers=[layer], initial_view_state=view_state)
r.show()

#r=pdk.Deck(layers=[layer],
           # map_style = 'mapbox://styles/mapbox/outdoors-v11',
#            mapbox_key='pk.eyJ1Ijoib25lZ3llb2wiLCJhIjoiY2xnZGdrNzQ1MDZjejNybnQ3azFsdDlkdiJ9.I46i27iqB-JeGqiVX51VqA',
#            initial_view_state=view_state)
#r.to_html('demo.html')
```
