## 단독주택 공동주택 현황

```python
import pandas as pd

df = pd.read_csv('단독_공동주택현황.csv', encoding='utf-8')
df
```

```python
df.columns
```

```python
del df['동별(1)'], df['동별(2)']
df
```

```python
dong = ['화양동','군자동','중곡1동','중곡2동','중곡3동','중곡4동', '능동', '구의1동','구의2동','구의3동', '광장동', '자양1동', '자양2동', '자양3동', '자양4동']
num_house = [1413, 1435, 1094, 1188, 1179, 1508, 786, 1070, 977, 424, 214, 1205, 848, 402, 1589]
num_vacanthouse=[3883, 3121, 2519, 3798, 2691, 5858, 1745, 4125, 6364,8646, 10135, 4193, 6218, 8180, 3024]

data = pd.DataFrame(dong, columns=['행정동'])
data['단독주택'] = num_house
data['공동주택'] = num_vacanthouse

data.to_csv('행정동별주택.csv', encoding='cp949', index=False)
```

```python
data1 = pd.read_csv('행정동별주택.csv', encoding='cp949', index_col=0)
data1
```

```python
data2 = data1.T
data2
```

```python
data2['구의동'] = data2['구의1동']+data2['구의2동']+data2['구의3동']
data2['자양동'] = data2['자양1동']+data2['자양2동']+data2['자양3동']+data2['자양4동']
data2['중곡동'] = data2['중곡1동']+data2['중곡2동']+data2['중곡3동']+data2['중곡4동']

del data2['구의1동'], data2['구의2동'], data2['구의3동'], data2['자양1동'], data2['자양2동'], data2['자양3동'], data2['자양4동'], data2['중곡1동'], data2['중곡2동'], data2['중곡3동'], data2['중곡4동']

data2
```

```python
data2 = data2[['중곡동', '군자동', '능동', '구의동', '광장동', '화양동', '자양동']]
data2
```

```python
data3 = data2.T
data3
```

```python
data3.to_csv('법정동별_주택.csv', encoding='cp949')
```

```python
import matplotlib.pyplot as plt
```
