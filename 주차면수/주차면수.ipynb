### 법정동별 주차면수 

```python
dong = ['중곡동', '군자동', '능동', '구의동', '광장동', '화양동', '자양동']
on_street_parkinglot = [672, 47, 162, 328, 188, 153, 845]
off_street_parkinglot = [16, 5, 4, 11, 2, 5, 19]
affiliated_parkinglot = [5541, 1074, 824, 3375, 336, 1138, 3535]

DF = pd.DataFrame(dong, columns=['법정동'])
DF['노상주차장'] = on_street_parkinglot
DF['노외주차장'] = off_street_parkinglot
DF['부설주차장'] = affiliated_parkinglot

DF
```

```python
DF['총 주차면수'] = DF['노상주차장'] + DF['노외주차장'] + DF['부설주차장']

DF
```

```python
DF.to_csv('법정동별_주차면수.csv', index=False, encoding='cp949')
```
