## 주차장 확보율
- (주차면수/차량수) * 100

```python
import pandas as pd
import numpy as np
```

```python
parking = pd.read_csv('법정동별_주차면수.csv', encoding='cp949')
parking
```

```python
car = pd.read_csv('법정동별_차량등록대수.csv', encoding='cp949')
car
```

```python
data = (parking['총 주차면수']/car['자동차등록대수'])*100
data
```

```python
dong = ['중곡동', '군자동', '능동', '구의동', '광장동', '화양동', '자양동']


df = pd.DataFrame(dong, columns=['법정동'])
df['주차장 확보율'] = (parking['총 주차면수']/car['자동차등록대수'])*100
df
```

```python
df.to_csv('법정동별_주차장확보율.csv', encoding='cp949', index=False)
df
```
