```python
data1 = pd.read_csv('법정동별_인구수.csv', encoding='cp949')
data1
```

```python
data1['광진구'] = data1['광진구'].str.replace(',','')
data1['중곡동'] = data1['중곡동'].str.replace(',','')
data1['군자동'] = data1['군자동'].str.replace(',','')
data1['능동'] = data1['능동'].str.replace(',','')
data1['구의동'] = data1['구의동'].str.replace(',','')
data1['광장동'] = data1['광장동'].str.replace(',','')
data1['화양동'] = data1['화양동'].str.replace(',','')
data1['자양동'] = data1['자양동'].str.replace(',','')
data1
```

```python
from prophet import Prophet 
 
dong_name = ['광진구', '중곡동', '군자동', '능동', '구의동', '광장동', '화양동', '자양동', 'Date']

future = list()
for i in range(2022, 2026): # 2022 2025
    for j in range(1, 13):
        future.append(['%04d-%02d' % (i,j)])

future = pd.DataFrame(future, columns = ['ds'])
future['ds']= pd.to_datetime(future['ds'])
```

```python
pd.to_datetime(data1['Date'], format='%Y%m')
```

```python
for i in range(len(data1['Date'])):
    date_time_str = data1['Date'][i]

    date = datetime.datetime.strptime(str(date_time_str), '%Y%m')

    data1['Date'][i] = date
```

```python
gwangjin_predict = pd.DataFrame()

for i,name in enumerate(dong_name):
    data2 = pd.DataFrame()
    data2 = data1[[name,'Date']]
    data2.columns = ['y','ds']
    data2['ds'] = pd.to_datetime(data2['ds'])

    model = Prophet().fit(data2)
#    print(i,"model_fit")

    forecast = model.predict(future)
    print(i,"model_predict")
    gwangjin_predict = pd.concat([gwangjin_predict, forecast['yhat']], ignore_index = True, axis=1)
```

```python
gwangjin_predict = pd.concat([gwangjin_predict, forecast['ds']], ignore_index = True, axis=1)

gwangjin_predict.columns = ['광진구', '중곡동', '군자동', '능동', '구의동', '광장동', '화양동', '자양동', 'Date']

gwangjin_predict=gwangjin_predict[12:]
```

```python
gwangjin_predict.to_csv('gwangjin_predict.csv',index = False, encoding='cp949')
```
