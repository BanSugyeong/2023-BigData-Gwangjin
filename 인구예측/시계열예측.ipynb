```python
from prophet import Prophet 
 
dong_name = ['중곡동', '군자동', '능동', '구의동', '광장동', '화양동', '자양동', 'Date']

future = list()
for i in range(2023, 2026): # 2022 2025
    for j in range(1, 13):
        future.append(['%04d-%02d' % (i,j)])

future = pd.DataFrame(future, columns = ['ds'])
future['ds']= pd.to_datetime(future['ds'])
```

```python
pd.to_datetime(data1['Date'], format='%Y%m')
```

```python
for i in range(len(data1['Date'])):
    date_time_str = data1['Date'][i]

    date = datetime.datetime.strptime(str(date_time_str), '%Y%m')

    data1['Date'][i] = date
```

```python
gwangjin_predict = pd.DataFrame()

for i,name in enumerate(dong_name):
    data2 = pd.DataFrame()
    data2 = data1[[name,'Date']]
    data2.columns = ['y','ds']
    data2['ds'] = pd.to_datetime(data2['ds'])

    model = Prophet().fit(data2)
#    print(i,"model_fit")

    forecast = model.predict(future)
    print(i,"model_predict")
    gwangjin_predict = pd.concat([gwangjin_predict, forecast['yhat']], ignore_index = True, axis=1)
```
