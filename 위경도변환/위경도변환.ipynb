### 23년 1월

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm
```

```python
df1 = pd.read_csv('불법주정차단속현황_2301.csv', encoding='cp949')
df1
```

```python
import googlemaps

googlemaps_key = "AIzaSyBEOAX2vHzQOdsm8ghf5a476a7LRe2Sng8"
gmaps = googlemaps.Client(key=googlemaps_key)
```

```python
def add_to_point(address):
    geo_location = gmaps.geocode(address)[0].get('geometry')
    
    lat = geo_location['location']['lat']
    lng = geo_location['location']['lng']
    
    return lat, lng
```

```python
gwangjin_points2301 = []
null_list2301 = []

for i in df1['단속장소']:
    address = i
    
    try:
        lat,lng = add_to_point(address)
        print(lat,lng)
        tmp_dic = {'단속장소':i,'위도':lat, '경도':lng}
        gwangjin_points2301.append(tmp_dic)
    except:
        null_list2301.append(i)
```

```python
df1_1 = pd.DataFrame(gwangjin_points2301)
df1_1
```

```python
null_list2301
```

```python
df1_1.to_csv('불법주정차단속현황_2301.csv', index=False, encoding='cp949')
df1_1
```
